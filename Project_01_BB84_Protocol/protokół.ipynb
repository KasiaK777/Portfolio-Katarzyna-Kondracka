```python
# importy
from qiskit import QuantumCircuit, execute, Aer
import numpy as np
import os

# funkcje pomocnicze
def generate_random_bases(num_bits):
    return np.random.randint(2, size=num_bits)

def apply_bb84_gate(circuit, bases, index):
    if bases[index] == 1:
        circuit.h(index)
    return circuit

# kod Anny
def anna_nadaje(circuit, num_qubits):
    anna_bases = generate_random_bases(num_qubits)
    
    for i in range(len(anna_bases)):
        circuit = apply_bb84_gate(circuit, anna_bases, i)
    
    os.environ['BAZY ANNA'] = "".join(anna_bases.astype(str))
    
    return circuit

# kod Eryka
def eryk_odbiera(circuit, num_qubits):
    eryk_bases = generate_random_bases(num_qubits)
    
    for i in range(len(eryk_bases)):
        circuit = apply_bb84_gate(circuit, eryk_bases, i)

    circuit.measure_all()
    key = execute(circuit.reverse_bits(), backend=Aer.get_backend('qasm_simulator'), shots=1).result().get_counts().most_frequent()
    print(f"Pełny klucz, który podsłuchał Eryk to:\t {key} \n")
    
    for i in range(len(eryk_bases)):
        circuit = apply_bb84_gate(circuit, eryk_bases, i)
    
    return circuit

# kod Bartka
def bartek_odbiera(circuit, num_qubits):
    bartek_bases = generate_random_bases(num_qubits)
    
    for i in range(len(bartek_bases)):
        circuit = apply_bb84_gate(circuit, bartek_bases, i)

    circuit.measure_all()

    key = execute(circuit.reverse_bits(), backend=Aer.get_backend('qasm_simulator'), shots=1).result().get_counts().most_frequent()
    
    os.environ['BAZY BARTEK'] = "".join(bartek_bases.astype(str))
    os.environ['STAN BARTEK'] = key
    
    return circuit

# kod konwersacji
def main():
    num_qubits = 20
    circuit = QuantumCircuit(num_qubits)

    circ = anna_nadaje(circuit, num_qubits)
    circ = eryk_odbiera(circ, num_qubits)
    circ = bartek_odbiera(circ, num_qubits)

    anna_bases = os.environ.get('BAZY ANNA')
    bartek_bases = os.environ.get('BAZY BARTEK')

    print(f"Bazy, których użyła Anna:\t {anna_bases}")
    print(f"Bazy, których użył Bartek:\t {bartek_bases} \n")

if __name__ == "__main__":
    main()
```
